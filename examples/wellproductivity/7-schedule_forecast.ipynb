```python
#Create the well object
name1 = 'well-3'
rte1 = 1515.78 # Rotary table Elevation
surf_coord1 = [1000000,1000000]#Point(1000100,1000000,520)
crs1 = 'EPSG:3117'

sched = {
    'initial':{
        'declination':dc.wor_declination(slope=2e-6, bsw_i=0.3,start_date=date(2021,1,1),end_date=date(2023,1,1),
                                         econ_limit=800,fluid_rate=2500),
    },
    'wo':{
        'declination':dc.wor_declination(slope=3e-6, bsw_i=0.3,start_date=date(2021,6,1),end_date=date(2025,1,1),
                                         econ_limit=10,fluid_rate=2700),
        'depend_start':True
    }
}


w3 = ph.well(name=name1, 
             rte=rte1, 
             surf_coord=surf_coord1, 
             crs=crs1,
            td = 9000,
             schedule = sched
            )




f3=w3.schedule_forecast(spread=False)
f3.head()
```

```python
sched1 = {
    'initialc1':{
        'declination':dc.wor_declination(slope=2e-6, bsw_i=0.3,start_date=date(2021,1,1),end_date=date(2023,1,1),
                                         econ_limit=1900,fluid_rate=5500),
        
    },
    'woc1':{
        'declination':dc.wor_declination(slope=3e-6, bsw_i=0.3,start_date=date(2021,6,1),end_date=date(2025,1,1),
                                         econ_limit=500,fluid_rate=5500),
        'depend_start':True,
        'time_delay':timedelta(days=160)
    }
}

sched2 = {
    'initialc2':{
        'declination':dc.wor_declination(slope=2e-6, bsw_i=0.3,start_date=date(2021,2,1),end_date=date(2023,1,1),
                                         econ_limit=900,fluid_rate=4500),
    },
    'woc2':{
        'declination':dc.wor_declination(slope=3e-6, bsw_i=0.3,start_date=date(2021,7,1),end_date=date(2025,1,1),
                                         econ_limit=500,fluid_rate=1700),
        'depend_start':True
    }
}


c1 = ph.well(name='c1', 
             rte=562, 
             surf_coord=[1000000,1000000], 
             crs='EPSG:3117',
            td = 9000,
             schedule = sched1
            )

c2 = ph.well(name='c2', 
             rte=562, 
             surf_coord=[1100000,1100000], 
             crs='EPSG:3117',
            td = 9000,
             schedule = sched2
            )

cg = ph.wells_group(c1,c2)


cg.describe()
```

```python
ff=cg.schedule_forecast()
ff
```

```python
fig, ax = plt.subplots()
sns.lineplot(x='time', y='qo', data=ff.reset_index(), hue='well', ax=ax)
ax.set_ylabel('time')
ax.set_xlim([date(2020,1,1),date(2025,1,1)])
```

```python
gg=ff.reset_index().groupby(['time'])['qo'].sum().reset_index()
sns.lineplot(x='time',y='qo',data=gg)
```

```python
jj=ff.reset_index().groupby(['well','time'])[['qw','qo']].sum().reset_index().pivot(index='time',columns='well',values=['qw','qo'])
jj.sum(axis=1,level=1)
```
