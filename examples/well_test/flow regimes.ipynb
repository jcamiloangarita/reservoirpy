# Infinte acting reservoir Flow

```python
bo=1.25
muo=1.5
ct=12e-6
ko=60
h=15
pi=4000
phi=0.15
rw=0.25
qo=300

r = np.array([0.25,5,10,50,100,500,1000])
t = np.array([0.01,0.1,1.10,12,24,36,48,72,90,1000])

rr,tt=np.meshgrid(r,t)

dd=wt.iarf_ei_dp(qo,bo,mu,ko,h,phi,ct,rr,tt)
pp = pi - dd
```

```python
fig, ax = plt.subplots(figsize=(9,5))

for i in pp:
    ax.plot(r,i)

ax.set_xlabel('radius')
ax.set_ylabel('pressure')
ax.set_xscale('log')
```
